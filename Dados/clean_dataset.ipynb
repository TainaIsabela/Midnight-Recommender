```python
import pandas as pd
import numpy as np
```

```python
# read the dataset

df = pd.read_csv('dados.csv')

df.head()
```

```python
df.isnull().sum()
```

```python
# drop the columns not needed for the recommender system

df = df.drop(['ISBN_13', 'ISBN_10', 'avaliacao', 'resenha', 'abandonos', 'relendo', 'querem_ler', 'lendo', 'leram', 'male', 'female', 'editora', 'idioma', 'descricao'], axis=1)
print(df.head())
df.isnull().sum()
```

```python
#drop the 2 rows with missing values

df = df.dropna()
df.isnull().sum()
```

```python
# now save the cleaned dataset

df.to_csv('livros-skoob.csv', index=False)
```
